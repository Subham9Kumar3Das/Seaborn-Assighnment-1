### Q-1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful Python data visualization library built on top of Matplotlib. Here are five common plots you can create using Seaborn, along with their primary uses:

Histograms:

Use: Display the distribution of a single numerical variable.
Description: Histograms group data into bins and show the frequency or count of data points in each bin, allowing you to visualize the data's distribution and central tendencies.
Scatter Plots:

Use: Explore the relationship between two numerical variables.
Description: Scatter plots display individual data points as dots on a two-dimensional plane. They help identify patterns, trends, and correlations between variables.
Box Plots (Box-and-Whisker Plots):

Use: Visualize the distribution and spread of numerical data, especially for multiple categories.
Description: Box plots display the summary statistics of a dataset, including the median, quartiles, and outliers. They are useful for comparing the distribution of data across different categories or groups.
Bar Plots (Bar Charts):

Use: Compare categorical data or show the relationship between categorical and numerical variables.
Description: Bar plots represent data with rectangular bars, where the length of each bar corresponds to the data's value. They are effective for displaying counts, frequencies, or aggregations for different categories.
Heatmaps:

Use: Visualize a matrix of data, often used for correlation matrices and categorical data analysis.
Description: Heatmaps use color to represent the magnitude of values in a matrix. 

### Q-2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('fmri')
```

```python
df.head()
```

```python
l=df['event'].unique()
l1=df['region'].unique()
```

```python
l,l1
```

```python
plt.figure(figsize=(10,4))
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Line Plot of FMRI Data for Different Events and Regions")


sns.lineplot(data=df,x='timepoint',y='signal',hue='event',style='region')
```

### Q-3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
df1=sns.load_dataset('titanic')
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title('Box Plot of Age by Pclass',color='red')
sns.boxplot(data=df1,x='pclass',y='age')
plt.subplot(1,2,2)
plt.title('Box Plot of Age by Fare',color='Green')
sns.boxplot(data=df1,x='pclass',y='fare')
plt.tight_layout()
plt.show()
```

### Q-4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
df2=sns.load_dataset('diamonds')
```

```python
df2.head()
```

```python
plt.figure(figsize=(12,6))
plt.title('Histogram Plot',color='g')
plt.xlabel('Price',color='red')
plt.ylabel('Magnitude',color='blue')
sns.histplot(data=df2,x='price',hue='cut',kde=True)
```

### Q-6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
df4=sns.load_dataset('flights')
```

```python
df4.head()
```

```python
df5=df4.pivot_table(index='month',columns='year',values='passengers')
```

```python
plt.figure(figsize=(12,4))
plt.title("Passenger Count Heatmap (Flights Dataset)")
sns.heatmap(df5,annot=True,fmt='d',cmap='BuPu')
```
